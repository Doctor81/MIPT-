```python
from tkinter import *
from tkinter import messagebox # нужно будет для сообщения ошибок
from tkinter import ttk

root = Tk() 
root.title("калькулятор")
#логика калькулятора
def calc(key):   #создал функцию на вход которой подается нажатая кнопка)
    global memory
    if key == "=": 
        str1 = "-+0123456789.*/" # тут перечислил то что мы можем ввести и исключил возможность написания букв
        if calc_entry.get() [0] not in str1:
            calc_entry.insert(END, "Первый символ не число")
            messagebox.showerror("Ошибка", "вы ввели не число")
        try:
            result = eval(calc_entry.get()) # добавил возможность счета
            calc_entry.insert(END, "=" + str(result)) 
        except:
            calc_entry.insert(END, "Ошибка") #избавился от ошибок
            messagebox.showerror("ошибка", "проверь правильность данных")
    elif key == "C":
       calc_entry.delete(0, END) #добавил возможность очистить поле

    elif key == "-/+":
        if "=" in calc_entry.get():
            calc_entry.delete(0, END)      #смена +/-
        try:
            if calc_entry.get() [0] == "-":
                calc_entry.delete(0)
            else:
                calc_entry.insert(0, "-")
        except IndexError:
            pass
    else:
        if "=" in calc_entry.get():
            calc_entry.delete(0,END)
        calc_entry.insert(END, key)
#создал кнопки
bttn_list = [
    "7", "8", "9", "+", "-",
    "4", "5", "6", "*","/",
    "1", "2", "3", "-/+", "=",
    "0", ".", "C", "", ""
]
# с помощью цикла for каждый элемент списка превратил в кнопку
r = 1 # строка 
c = 0 # столбец
for i in bttn_list:
    rel = ""
    cmd=lambda x = i: calc(x)
    ttk.Button(root, text=i, command=cmd).grid(row=r,column=c) 
    c += 1
    if c>4:
        c = 0
        r += 1

calc_entry = Entry(root, width=33) #создал окно для ввода
calc_entry.grid(row=0,column=0, columnspan=5)# выбрал место для этого окна, разместил в нулевой строке и нулевом столбце

root.mainloop()
```
